```python
import pandas
unrate=pandas.read_csv("UNRATE.csv")
unrate


```

```python
unrate["DATE"]=pandas.to_datetime(unrate["DATE"])
unrate
```

```python
import matplotlib.pyplot as plt 
plt.plot(unrate["DATE"].head(12),unrate["VALUE"].head(12),color='red',ls='--',linewidth=2)
plt.xlabel("Month",color='Blue')
plt .ylabel("Unemployment Rate",color='Blue')
plt.title("Monthly Unemployment Trends, 1948")
plt.xticks(rotation=45)
plt.show()
```

```python
### Checking seasonality across multiple years
fig=plt.figure(figsize=(12,12))
ax1=fig.add_subplot(5,1,1)
ax2=fig.add_subplot(5,1,2)
ax3=fig.add_subplot(5,1,3)
ax4=fig.add_subplot(5,1,4)
ax5=fig.add_subplot(5,1,5)
ax1.plot(unrate[0:12]['DATE'], unrate[0:12]['VALUE'])
ax2.plot(unrate[12:24]['DATE'], unrate[12:24]['VALUE'])
ax3.plot(unrate[24:36]['DATE'], unrate[24:36]['VALUE'])
ax4.plot(unrate[36:48]['DATE'], unrate[36:48]['VALUE'])
ax5.plot(unrate[48:60]['DATE'], unrate[48:60]['VALUE'])


plt.show()
```

```python
### Multiple years in single plot 

unrate['MONTH'] = unrate['DATE'].dt.month
fig=plt.figure(figsize=(6,3))
plt.plot(unrate['MONTH'][0:12],unrate['VALUE'][0:12],color='red')
plt.plot(unrate['MONTH'][12:24],unrate['VALUE'][12:24],color='blue')
plt.show()
```

```python
### Five years in one column
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i])
    
plt.show()
```

```python
### Adding legends and label
fig = plt.figure(figsize=(10,6))
ax=fig.add_subplot(1,1,1)
colors = ['red', 'blue', 'green', 'orange', 'black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label=str(1948+i)
    ax.plot(subset['MONTH'], subset['VALUE'], c=colors[i],label=label)
ax.legend(loc="upper left")
ax.set(title="Monthly Unemployment Trends, 1948-1952",ylabel="Unemployment Rate, Percent",xlabel="Month, Integer")



plt.show()
```
